```python
import matplotlib.pyplot as plt

df.hist(bins=20, figsize=(15, 10), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Numerical Features', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

sns.pairplot(df, hue='Survived')
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Plot histograms for key numerical columns
df[['Age', 'Fare', 'Survived']].hist(bins=20, figsize=(12, 6), color='skyblue', edgecolor='black')
plt.suptitle('Histogram of Age, Fare, and Survival', fontsize=16)
plt.tight_layout()
plt.show()
```

#Summary

- The dataset contains information on Titanic passengers including demographics, ticket class, fare, and survival.
- Age and Fare columns showed some skewness.
- Correlation heatmap revealed that Fare and Pclass were moderately correlated with survival.
- Female passengers had a significantly higher survival rate than males.
- Passengers in 1st class had better chances of survival.
- Missing values were present in Age and Cabin columns and were handled/observed carefully.
